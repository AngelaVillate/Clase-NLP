```python
import kagglehub
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import colorcet as cc
import os
```

```python
# Descargar dataset
dataset_path = kagglehub.dataset_download("isabelocastillo/jugadores-de-futbol-de-la-rfef-temporada-2023-2024")
print("Path to dataset files:", dataset_path)
```

```python
# List the files within the dataset path
files_in_dataset_path = os.listdir(dataset_path)
print("Files in dataset path:", files_in_dataset_path)
```

```python
# Find the CSV file
csv_file = next((file for file in files_in_dataset_path if file.endswith(".csv")), None)

if csv_file:
    # Construct the correct path to the CSV file
    csv_file_path = os.path.join(dataset_path, csv_file)

    # Cargar datos
    data = pd.read_csv(csv_file_path)  # Read the CSV file using the correct path
else:
    print("CSV file not found in the dataset path.")
```

```python
# Seleccionar columnas relevantes para el análisis
columns_to_use = ['id', 'nombre', 'partidos_jugados', 'partidos_titular', 'goles', 'asistencias', 'tarjeta_amarilla', 'edad', 'altura', 'valor', 'rating', 'posicion']
data = data[columns_to_use]
```

```python
# Convertir 'valor' a numérico
data['valor'] = data['valor'].replace('[^0-9]', '', regex=True).astype(float)
```

```python
# Ejemplo de visualización con paleta monocromática
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='edad', kde=True, color='darkblue')
plt.title("Distribución de edad de los jugadores (monocromático)")
plt.show()
```

```python
# Visualización con paleta tríada
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='posicion', y='goles', palette=sns.color_palette("Set1"))
plt.xticks(rotation=90)
plt.title("Goles por posición (paleta tríada)")
plt.show()
```

5. ¿Esto es una paleta triada? Justifica tu respuesta

```python
# Agrupar jugadores en tres categorías de edad
bins = [15, 23, 30, 40]  # Definir los rangos de edad
labels = ["Jóvenes (≤23)", "Adultos (24-30)", "Veteranos (≥31)"]
data["grupo_edad"] = pd.cut(data["edad"], bins=bins, labels=labels, include_lowest=True)

# Visualización con paleta tríada mejorada
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='grupo_edad', y='goles', palette=sns.color_palette("Set1"))
plt.title("Goles por Grupo de Edad (Paleta Tríada)")
plt.show()
```

```python
# Visualización con paleta divergente
plt.figure(figsize=(10, 6))
sns.heatmap(data.pivot_table(index='posicion', values='asistencias', aggfunc='mean').sort_values('asistencias'), cmap='RdBu', annot=True)
plt.title("Mapa de calor de asistencias por posición (paleta divergente)")
plt.show()
```

```python
# Visualización con paleta de degradado mejorada
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='partidos_titular', y='rating', cmap='magma', fill=True, levels=30, alpha=0.8)
plt.title("Densidad de Partidos como Titular vs Rating (paleta de degradado mejorada)")
plt.show()
```

Colores y significado:

La paleta de colores "magma" (degradado) indica densidad.
Zonas más claras: Representan regiones donde hay mayor concentración de jugadores con esos valores de titularidad y rating.
Zonas oscuras: Representan menos jugadores con esa combinación de valores.
