```python
from IPython import get_ipython
from IPython.display import display
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import spacy
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA, LatentDirichletAllocation
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline
import pyLDAvis
```

```python
# Download the Spanish language model for spaCy
!python -m spacy download es_core_news_sm
# Cargar modelo de spaCy en español
nlp = spacy.load("es_core_news_sm")
nltk.download("stopwords")
stop_words = set(stopwords.words("spanish"))
```

```python
# 1. Web Scraping de Wikipedia
def obtener_descripciones():
    cientificas = ["Marie_Curie", "Rosalind_Franklin", "Ada_Lovelace", "Dorothy_Crowfoot_Hodgkin", "Barbara_McClintock",
                   "Jane_Goodall", "Lise_Meitner", "Rachel_Carson", "Vera_Rubin", "Emmy_Noether", "Hypatia", "Mary_Anning",
                   "Caroline_Herschel", "Maria_Gaetana_Agnesi", "Chien-Shiung_Wu", "Sophie_Germain", "Grace_Hopper",
                   "Henrietta_Swan_Leavitt", "Maria_Mitchell", "Jocelyn_Bell_Burnell"]
    base_url = "https://es.wikipedia.org/wiki/"
    descripciones = {}

    for cientifica in cientificas:
        url = base_url + cientifica
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        parrafos = soup.find_all("p")
        texto = " ".join([p.text for p in parrafos[:5]])
        descripciones[cientifica] = texto

    return descripciones
```

```python
# prompt: GENERA CÓDIGO PARA LIMPIAR TEXTO UNICODE, STOPWORDS EN ESPAÑOL, MINUSCULAS, ELIMINAR NÚMEROS, SÍMBOLOS EXTRAÑOS TODO LO EXTRAÑO

import re

def limpiar_texto(texto):
    """
    Limpia un texto eliminando caracteres especiales, números, stopwords,
    convirtiendo a minúsculas y normalizando el texto Unicode.
    """
    # Normalización Unicode
    texto = texto.lower()  # Convertir a minúsculas

    # Eliminar números
    texto = re.sub(r'\d+', '', texto)

    # Eliminar símbolos extraños y caracteres no alfanuméricos
    texto = re.sub(r'[^\w\s]', '', texto)

    # Eliminar espacios en blanco múltiples
    texto = re.sub(r'\s+', ' ', texto).strip()


    # Eliminar stopwords en español
    texto = " ".join([palabra for palabra in texto.split() if palabra not in stop_words])

    return texto

# Ejemplo de uso con el diccionario 'descripciones'
descripciones_limpias = {}
for cientifica, descripcion in descripciones.items():
    descripciones_limpias[cientifica] = limpiar_texto(descripcion)

descripciones_limpias
```

```python
# prompt: GENERA CÓDIGO PARA TOKENIZAR Y LEMATIZAR CON STEAMING (RECUERDA ES TEXTO EN ESPAÑOL)

def tokenizar_lematizar(texto):
    """
    Tokeniza y lematiza un texto en español utilizando spaCy.
    """
    doc = nlp(texto)
    tokens_lemas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return tokens_lemas

# Ejemplo de uso con las descripciones limpias
descripciones_tokenizadas_lemmatizadas = {}
for cientifica, descripcion in descripciones_limpias.items():
    descripciones_tokenizadas_lemmatizadas[cientifica] = tokenizar_lematizar(descripcion)

descripciones_tokenizadas_lemmatizadas
```

```python
# vectoriza con tf idf

from sklearn.feature_extraction.text import TfidfVectorizer

def vectorizar_tfidf(descripciones_limpias):
    """
    Vectoriza las descripciones limpias utilizando TF-IDF.
    """
    vectorizador = TfidfVectorizer()
    matriz_tfidf = vectorizador.fit_transform(descripciones_limpias.values())
    return matriz_tfidf, vectorizador

# Ejemplo de uso
matriz_tfidf, vectorizador = vectorizar_tfidf(descripciones_limpias)

# Imprimir la matriz TF-IDF (opcional)
print(matriz_tfidf.toarray())

# Mostrar las palabras clave del vocabulario (opcional)
print(vectorizador.get_feature_names_out())
```

```python
#  distancia del coseno a partir de la vectorización tf idf

def calcular_distancia_coseno(matriz_tfidf):
    """
    Calcula la distancia del coseno entre todos los vectores de la matriz TF-IDF.
    """
    distancias_coseno = 1 - cosine_similarity(matriz_tfidf)
    return distancias_coseno

# Ejemplo de uso
distancias = calcular_distancia_coseno(matriz_tfidf)

# Imprimir la matriz de distancias del coseno
distancias
```

```python
# matriz de cercania con colorimetria entre purpura y amarillo a partir de la distancia del cosen en la que se comparan todas las autoras

import matplotlib.pyplot as plt
import seaborn as sns

def generar_matriz_cercania(distancias, descripciones_limpias):
    """
    Genera una matriz de cercanía con colorimetría entre púrpura y amarillo.
    """
    # Crear el mapa de calor
    plt.figure(figsize=(14, 8))  # Ajustar el tamaño de la figura
    sns.heatmap(distancias,
                annot=True,
                cmap="plasma_r",  # Usar un mapa de colores púrpura-amarillo
                xticklabels=descripciones_limpias.keys(),
                yticklabels=descripciones_limpias.keys(),
                vmin=0, vmax=1) #valores minimos y maximos
    plt.title("Matriz de Cercanía entre Autoras")
    plt.xlabel("Autoras")
    plt.ylabel("Autoras")
    plt.xticks(rotation=45, ha="right")  # Rotar etiquetas del eje x
    plt.yticks(rotation=0)
    plt.tight_layout()  # Ajustar el diseño para evitar superposiciones
    plt.show()


generar_matriz_cercania(distancias, descripciones_limpias)
```

```python
# genera pca y visualiza

import matplotlib.pyplot as plt
def realizar_pca(matriz_tfidf, n_components=2):
    """
    Realiza el análisis de componentes principales (PCA) en la matriz TF-IDF.
    """
    pca = PCA(n_components=n_components)
    componentes_principales = pca.fit_transform(matriz_tfidf.toarray())
    return componentes_principales, pca

# Ejemplo de uso
componentes_principales, pca = realizar_pca(matriz_tfidf)

def visualizar_pca(componentes_principales, descripciones_limpias):
    """
    Visualiza los resultados del PCA en un gráfico de dispersión.
    """
    plt.figure(figsize=(10, 6))
    for i, cientifica in enumerate(descripciones_limpias.keys()):
        plt.scatter(componentes_principales[i, 0], componentes_principales[i, 1])
        plt.annotate(cientifica, (componentes_principales[i, 0], componentes_principales[i, 1]))
    plt.xlabel("Componente Principal 1")
    plt.ylabel("Componente Principal 2")
    plt.title("Visualización PCA de las descripciones de científicas")
    plt.grid(True)
    plt.show()

# Ejemplo de uso
visualizar_pca(componentes_principales, descripciones_limpias)
```

```python
# LDA topic modeling
import pyLDAvis.gensim_models
import gensim
from IPython.display import IFrame

def realizar_lda(matriz_tfidf, num_topics=5):
    """
    Realiza el modelado de temas LDA en la matriz TF-IDF.
    """
    lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
    lda.fit(matriz_tfidf)
    return lda

# Ejemplo de uso
lda_model = realizar_lda(matriz_tfidf)

def visualizar_lda(lda_model, vectorizador, descripciones_limpias):
  """Visualiza los temas LDA usando pyLDAvis."""
  # Convertir la matriz TF-IDF a un formato compatible con gensim
  # corpus = matriz_tfidf.toarray().astype(np.float64)  # Convertir a float64
  # vectorizador ya se debe tener del preprocesamiento anterior
  # Crear un diccionario a partir de los feature names del vectorizador
  id2word = {i: word for i, word in enumerate(vectorizador.get_feature_names_out())}

  # Create a Gensim Dictionary from your existing id2word:
  dictionary = gensim.corpora.Dictionary.from_corpus(gensim.matutils.Sparse2Corpus(matriz_tfidf, documents_columns=False), id2word=id2word)

  # Convertir el corpus a un objeto gensim.corpora.MmCorpus
  # gensim_corpus = [gensim.matutils.Dense2vec(doc) for doc in corpus] # This line was causing the error
  gensim_corpus = gensim.matutils.Sparse2Corpus(matriz_tfidf, documents_columns=False) # Use Sparse2Corpus instead

  # Convertir el modelo LDA a un objeto gensim.models.LdaModel
  # num_topics, id2word, eta, alpha deben coincidir con los del modelo original de sklearn
  gensim_lda_model = gensim.models.LdaModel(
      corpus=gensim_corpus,
      id2word=id2word,
      num_topics=lda_model.n_components,
      eta=lda_model.components_ / lda_model.components_.sum(axis=1)[:, np.newaxis],
      alpha=lda_model.doc_topic_prior_,
      # random_state=42  # Add this line to ensure consistent results (optional)
      passes=20  # Increase passes for better convergence (optional)
  )
  # gensim_lda_model.state.sstats = lda_model.components_ # Copiar los sstats del modelo de sklearn

  # Pasar el diccionario Gensim a pyLDAvis.gensim_models.prepare
  vis_data = pyLDAvis.gensim_models.prepare(gensim_lda_model, gensim_corpus, dictionary)

  # Embed the visualization using IFrame
  pyLDAvis.save_html(vis_data, 'lda_visualization.html')
  return IFrame(src='./lda_visualization.html', width=1000, height=800)


# Ejemplo de uso
display(visualizar_lda(lda_model, vectorizador, descripciones_limpias))
```
