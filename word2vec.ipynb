```python
# Paso 2: Importar bibliotecas
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Paso 3: Preparar el corpus en espa침ol
corpus = [
    "El gato est치 sobre la mesa",
    "El perro juega en el parque",
    "La comida est치 muy sabrosa",
    "El ni침o corre hacia la escuela",
    "La flor es muy bonita",
    "El sol brilla en el cielo",
    "La luna ilumina la noche",
    "El p치jaro canta en el 치rbol",
    "El coche circula por la carretera",
    "El libro est치 sobre la estanter칤a"
]

# Paso 4: Preprocesar el corpus (convertir a min칰sculas y dividir en palabras)
processed_corpus = [sentence.lower().split() for sentence in corpus]
print("Corpus preprocesado:", processed_corpus)

# Paso 5: Entrenar el modelo CBOW
model = Word2Vec(sentences=processed_corpus, vector_size=100, window=2, min_count=1, sg=0)
# sg=0 indica que usamos CBOW (si sg=1, ser칤a Skip-gram)

# Paso 6: Obtener los vectores de palabras y sus palabras correspondientes
words = list(model.wv.index_to_key)  # Lista de palabras en el vocabulario
vectors = [model.wv[word] for word in words]  # Vectores de palabras

# Paso 7: Reducir la dimensionalidad a 2D usando PCA
pca = PCA(n_components=2)  # Reducir a 2 dimensiones
vectors_2d = pca.fit_transform(vectors)  # Aplicar PCA a los vectores

# Paso 8: Visualizar los vectores en un gr치fico 2D
plt.figure(figsize=(10, 8))  # Tama침o del gr치fico
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], alpha=0.5)  # Graficar puntos

# A침adir etiquetas a los puntos
for i, word in enumerate(words):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]))

# A침adir t칤tulo y mostrar el gr치fico
plt.title("Visualizaci칩n de vectores de palabras usando PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()
```

```python
from gensim.models import Word2Vec

# Define tu corpus de texto
corpus = [
    "El gato est치 sobre la mesa",
    "El perro juega en el parque",
    "La comida est치 muy sabrosa",
    "El ni침o corre hacia la escuela",
    "La flor es muy bonita"
]

```

```python
# Preprocesa el corpus (tokenizaci칩n)
processed_corpus = [sentence.lower().split() for sentence in corpus]

# Crea y entrena el modelo CBOW
model = Word2Vec(
    sentences=processed_corpus,
    vector_size=20,  # Dimensi칩n de los vectores de palabras
    window=2,  # Tama침o de la ventana de contexto
    min_count=1,  # Frecuencia m칤nima de palabras para incluirlas en el vocabulario
    sg=0,  # 0 para CBOW, 1 para Skip-gram
)
```

```python
# Accede a los vectores de palabras
vector_gato = model.wv['gato']
print(vector_gato)  # Imprime el vector de la palabra "gato"
```

```python

# Encuentra palabras similares
similar_words = model.wv.most_similar('gato', topn=5)
print(similar_words)  # Imprime las 5 palabras m치s similares a "gato"
```

```python
import re
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from gensim.models import FastText
```

# ==============================
# EXPLICACI칍N DE FASTTEXT
# ==============================
# FastText es un modelo desarrollado por Facebook AI Research que extiende Word2Vec.
# En lugar de tratar cada palabra como un 칰nico token, descompone las palabras en n-gramas,
# lo que permite capturar mejor la morfolog칤a de las palabras y manejar palabras fuera de vocabulario.
# Es especialmente 칰til en idiomas con muchas variaciones morfol칩gicas, como el espa침ol.

```python
# 1. Carga y preprocesamiento del corpus
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    return tokens
```

```python
corpus_text = [
    "El tribunal resolvi칩 el caso con base en la jurisprudencia establecida.",
    "La corte constitucional emiti칩 una sentencia hist칩rica sobre derechos humanos.",
    "El abogado defensor present칩 un recurso de apelaci칩n contra la decisi칩n.",
    "El juez dict칩 una medida cautelar para proteger los intereses de las partes.",
    "Los demandantes solicitaron una indemnizaci칩n por da침os y perjuicios."
]

processed_corpus = [preprocess_text(text) for text in corpus_text]
```

```python
# 2. Creaci칩n de modelo FastText
# FastText es capaz de manejar palabras fuera de vocabulario mejor que Word2Vec.
model = FastText(sentences=processed_corpus, vector_size=50, window=5, min_count=1, workers=4, epochs=20)
```

```python
# 3. Medici칩n y visualizaci칩n de vectores
def plot_words(model, words):
    vectors = np.array([model.wv[word] for word in words if word in model.wv])
    # The perplexity should be less than the number of samples (words in this case)
    # Since you have 6 words, set perplexity to a value less than 6, e.g., 5 or lower
    tsne = TSNE(n_components=2, random_state=42, perplexity=5)
    reduced_vectors = tsne.fit_transform(vectors)

    plt.figure(figsize=(8, 6))
    for word, (x, y) in zip(words, reduced_vectors):
        plt.scatter(x, y)
        plt.annotate(word, (x, y), fontsize=12)
    plt.show()
```

```python
words_to_plot = ['tribunal', 'corte', 'sentencia', 'juez', 'abogado', 'demanda']
plot_words(model, words_to_plot)
```
