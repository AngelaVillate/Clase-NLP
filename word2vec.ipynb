```python
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# Corpus de ejemplo en español
corpus = [
    "El gato está sobre la mesa",
    "El perro juega en el parque",
    "La comida está muy sabrosa",
    "El niño corre hacia la escuela",
    "La flor es muy bonita"
]
```

```python
# Preprocesar el corpus
processed_corpus = [sentence.lower().split() for sentence in corpus]
```

```python
# Entrenar el modelo CBOW
model = Word2Vec(sentences=processed_corpus, vector_size=100, window=2, min_count=1, sg=0)
```

```python
# Obtener los vectores de palabras y sus palabras correspondientes
words = list(model.wv.index_to_key)
vectors = [model.wv[word] for word in words]
```

```python
# Reducir a 2 dimensiones con t-SNE
import numpy as np
tsne = TSNE(n_components=2, random_state=0, perplexity=5) # Set perplexity to 5 (or any value less than 20)
vectors_2d = tsne.fit_transform(np.array(vectors)) # Convert 'vectors' to a numpy array
```

```python
# Graficar
plt.figure(figsize=(10, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], alpha=0.5)
```

```python
# Paso 2: Importar bibliotecas
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Paso 3: Preparar el corpus en español
corpus = [
    "El gato está sobre la mesa",
    "El perro juega en el parque",
    "La comida está muy sabrosa",
    "El niño corre hacia la escuela",
    "La flor es muy bonita",
    "El sol brilla en el cielo",
    "La luna ilumina la noche",
    "El pájaro canta en el árbol",
    "El coche circula por la carretera",
    "El libro está sobre la estantería"
]

# Paso 4: Preprocesar el corpus (convertir a minúsculas y dividir en palabras)
processed_corpus = [sentence.lower().split() for sentence in corpus]
print("Corpus preprocesado:", processed_corpus)

# Paso 5: Entrenar el modelo CBOW
model = Word2Vec(sentences=processed_corpus, vector_size=100, window=2, min_count=1, sg=0)
# sg=0 indica que usamos CBOW (si sg=1, sería Skip-gram)

# Paso 6: Obtener los vectores de palabras y sus palabras correspondientes
words = list(model.wv.index_to_key)  # Lista de palabras en el vocabulario
vectors = [model.wv[word] for word in words]  # Vectores de palabras

# Paso 7: Reducir la dimensionalidad a 2D usando PCA
pca = PCA(n_components=2)  # Reducir a 2 dimensiones
vectors_2d = pca.fit_transform(vectors)  # Aplicar PCA a los vectores

# Paso 8: Visualizar los vectores en un gráfico 2D
plt.figure(figsize=(10, 8))  # Tamaño del gráfico
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], alpha=0.5)  # Graficar puntos

# Añadir etiquetas a los puntos
for i, word in enumerate(words):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]))

# Añadir título y mostrar el gráfico
plt.title("Visualización de vectores de palabras usando PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()
```

```python
from gensim.models import Word2Vec

# Define tu corpus de texto
corpus = [
    "El gato está sobre la mesa",
    "El perro juega en el parque",
    "La comida está muy sabrosa",
    "El niño corre hacia la escuela",
    "La flor es muy bonita"
]

```

```python
# Preprocesa el corpus (tokenización)
processed_corpus = [sentence.lower().split() for sentence in corpus]

# Crea y entrena el modelo CBOW
model = Word2Vec(
    sentences=processed_corpus,
    vector_size=20,  # Dimensión de los vectores de palabras
    window=2,  # Tamaño de la ventana de contexto
    min_count=1,  # Frecuencia mínima de palabras para incluirlas en el vocabulario
    sg=0,  # 0 para CBOW, 1 para Skip-gram
)
```

```python
# Accede a los vectores de palabras
vector_gato = model.wv['gato']
print(vector_gato)  # Imprime el vector de la palabra "gato"
```

```python

# Encuentra palabras similares
similar_words = model.wv.most_similar('gato', topn=5)
print(similar_words)  # Imprime las 5 palabras más similares a "gato"
```

```python
import re
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from gensim.models import FastText
```

# ==============================
# EXPLICACIÓN DE FASTTEXT
# ==============================
# FastText es un modelo desarrollado por Facebook AI Research que extiende Word2Vec.
# En lugar de tratar cada palabra como un único token, descompone las palabras en n-gramas,
# lo que permite capturar mejor la morfología de las palabras y manejar palabras fuera de vocabulario.
# Es especialmente útil en idiomas con muchas variaciones morfológicas, como el español.

```python
# 1. Carga y preprocesamiento del corpus
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = text.split()
    return tokens
```

```python
corpus_text = [
    "El tribunal resolvió el caso con base en la jurisprudencia establecida.",
    "La corte constitucional emitió una sentencia histórica sobre derechos humanos.",
    "El abogado defensor presentó un recurso de apelación contra la decisión.",
    "El juez dictó una medida cautelar para proteger los intereses de las partes.",
    "Los demandantes solicitaron una indemnización por daños y perjuicios."
]

processed_corpus = [preprocess_text(text) for text in corpus_text]
```

```python
# 2. Creación de modelo FastText
# FastText es capaz de manejar palabras fuera de vocabulario mejor que Word2Vec.
model = FastText(sentences=processed_corpus, vector_size=50, window=5, min_count=1, workers=4, epochs=20)
```

```python
# 3. Medición y visualización de vectores
def plot_words(model, words):
    vectors = np.array([model.wv[word] for word in words if word in model.wv])
    # The perplexity should be less than the number of samples (words in this case)
    # Since you have 6 words, set perplexity to a value less than 6, e.g., 5 or lower
    tsne = TSNE(n_components=2, random_state=42, perplexity=5)
    reduced_vectors = tsne.fit_transform(vectors)

    plt.figure(figsize=(8, 6))
    for word, (x, y) in zip(words, reduced_vectors):
        plt.scatter(x, y)
        plt.annotate(word, (x, y), fontsize=12)
    plt.show()
```

```python
words_to_plot = ['tribunal', 'corte', 'sentencia', 'juez', 'abogado', 'demanda']
plot_words(model, words_to_plot)
```
