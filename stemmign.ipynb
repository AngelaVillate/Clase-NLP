```python
import re
import spacy
import fitz  # PyMuPDF
from nltk.stem.snowball import SnowballStemmer
import unicodedata
```

```python
# Cargar modelo de spaCy en español
nlp = spacy.load("es_core_news_sm")
```

```python
# Inicializar el stemmer para español
stemmer = SnowballStemmer("spanish")
```

```python
# Ruta del archivo PDF
pdf_path = "/content/certificado (2).pdf"
```

```python
# Función para leer el texto del PDF
def read_pdf(file_path):
    text = ""
    doc = fitz.open(file_path)
    for page in doc:
        text += page.get_text()
    doc.close()
    return text
```

```python
# Preprocesamiento del texto
def preprocess_text(text):
    # Normalizar a Unicode y convertir a minúsculas
    text = unicodedata.normalize('NFKD', text)
    text = text.lower()
    text = re.sub(r"[^\w\sáéíóúñü]", "", text)

    # Tokenización
    tokens = text.split()

    # Lematización
    doc = nlp(text)
    lemmatized_tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

    # Stemming
    stemmed_tokens = [stemmer.stem(token) for token in lemmatized_tokens]

    return " ".join(stemmed_tokens), tokens
```

```python
# Leer, preprocesar y mostrar el resultado
pdf_text = read_pdf(pdf_path)
processed_text, _ = preprocess_text(pdf_text) # Unpack the tuple, ignore the tokens
print(processed_text)

# Exportar a un archivo TXT
with open("processed_text.txt", "w", encoding="utf-8") as f:
    f.write(processed_text)
    print("Texto procesado exportado a processed_text.txt")
```

```python
from collections import Counter
```

 Se utiliza la clase Counter para contar la frecuencia de cada token en la lista processed_tokens. El resultado se almacena en un diccionario llamado word_frequencies donde las claves son los tokens y los valores son sus frecuencias.

```python
!pip install pandas
import pandas as pd
from collections import Counter

# Leer, preprocesar y mostrar el resultado
pdf_text = read_pdf(pdf_path)
processed_text, _ = preprocess_text(pdf_text)  # Unpack the tuple, ignore the tokens

# Obtener los tokens procesados (stemmed)
processed_tokens = processed_text.split()

# Calcular la frecuencia de las palabras
word_frequencies = Counter(processed_tokens)

# Crear un DataFrame de Pandas
df = pd.DataFrame(word_frequencies.items(), columns=["Palabra (Stemmed)", "Frecuencia"])
df = df.sort_values(by=["Frecuencia"], ascending=False)

# Mostrar la tabla
display(df)
```
