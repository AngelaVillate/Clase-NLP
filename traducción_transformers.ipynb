```python
# Importar las bibliotecas
from transformers import MarianMTModel, MarianTokenizer
```

```python
# Cargar el modelo y el tokenizer preentrenados
modelo_nombre = "Helsinki-NLP/opus-mt-es-en"
tokenizer = MarianTokenizer.from_pretrained(modelo_nombre)
modelo = MarianMTModel.from_pretrained(modelo_nombre)
```

```python
# Función para traducir texto
def traducir(texto):
    # Tokenizar el texto
    tokens = tokenizer(texto, return_tensors="pt", padding=True, truncation=True)

    # Generar la traducción
    traduccion_ids = modelo.generate(**tokens)

    # Decodificar la traducción
    traduccion = tokenizer.batch_decode(traduccion_ids, skip_special_tokens=True)

    return traduccion[0]
```

```python
# Ejemplo de uso
texto_espanol = "Hola, la IA se ha desarrollado gracias a los avances del procesamiento del lenguaje natural"
texto_ingles = traducir(texto_espanol)

print(f"Texto en español: {texto_espanol}")
print(f"Traducción en inglés: {texto_ingles}")
```
